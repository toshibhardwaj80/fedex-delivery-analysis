```python
# Importing matplotlib for visualization
import matplotlib.pyplot as plt

# Bar chart for shipment mode vs average cost
df.groupby('Shipment Mode')['Line Item Value'].mean().plot(kind='bar')

# Displaying the graph
plt.show()
```

```python
# Pie chart
df['Shipment Mode'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
# Histogram
df['Line Item Value'].hist()
plt.show()
```

### Insights

- Air shipment mode has higher average cost compared to other shipment modes.
- Certain vendors have significantly higher average costs, indicating cost variation across suppliers.
- Shipment mode plays a crucial role in determining overall cost.
- Some countries show higher delivery expenses, possibly due to logistics and distance factors.
- Most shipments are done using a particular shipment mode, showing its popularity.

### Conclusion

This analysis shows that shipment mode and vendor significantly impact delivery cost and efficiency. By optimizing shipment methods and selecting efficient vendors, companies can reduce costs and improve supply chain performance.
